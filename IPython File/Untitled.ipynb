```python
# TODO: Minimum price of the data
minimum_price = np.min(target)

# TODO: Maximum price of the data
maximum_price = np.max(target)

# TODO: Mean price of the data
mean_price = np.mean(target)

# TODO: Median price of the data
median_price = np.median(target)

# TODO: Standard deviation of prices of the data
std_price = np.std(target)

# Show the calculated statistics
print "Statistics for  dataset:\n"
print "Minimum redshift: {:,.2f}".format(minimum_price)
print "Maximum redshift: {:,.2f}".format(maximum_price)
print "Mean redshift: {:,.2f}".format(mean_price)
print "Median redshift {:,.2f}".format(median_price)
print "Standard deviation of redshift: ${:,.2f}".format(std_price)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=42)

# Success
print "Training and testing split was successful."
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X_train,y_train)
regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

```python
score = performance_metric(y_test,regressor.predict(X_test) )
print "Model has a coefficient of determination, R^2, of {:.3f}.".format(score)
```
